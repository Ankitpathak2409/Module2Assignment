# Q-1 Explain the main components of client-server architecture and their functions?
- Client-server architecture is a computing model where multiple client(user) request a servicce from a server and then the server processes the request send by the client and sent a optimized response to the client.It is mainly used in web application, networking and database management system.
1. **Client:-** A Client refer to any software application or devices that request services or resource from a server.
  
  **Function:-**
    * Sends requests to the server (e.g., web browser requesting a
webpage).
    * Process and present recieved data to the user.

2. **Server:-** The server is a powerful computer or software application that processes the client request and provides the required services and resources.

    **Function:-**
      * Store, manages and process data.
      * Responds to multiple client requests simultaneously.

3. **Network:-**The network is the communication channel that connects the client and server, allowing data exchange.
  
    **Function:-**
      * Enables seamless data transmission using protocols like HTTP, HTTPS, TCP/IP.
      * Can be wired (Ethernet) or wireless (Wi-Fi, mobile networks).

4. **Protocols:-**Protocols are rules and standards that govern communication between the client and the server.
    **Function:-**
      * Define how requests and responses are structured.
      * Examples include:

         HTTP/HTTPS (for web communication).

         FTP (for file transfers).

         SMTP/IMAP/POP3 (for emails).

         SQL (for database queries).

5. **Database:-** The database stores and manages structured data that the server retrieves and manipulates as per the client requests.
    
     **Function:-**
        * Stores data persistently.
        * Allows efficient data retrieval and management using SQL or NoSQL databases.

#Q-3 Describe the role of a web browser's developer tools in web development?

 * Web browsers come equipped with Developer Tools (DevTools) that help developers inspect, debug, and optimize websites. These tools provide a set of features that make it easier to understand and refine how a web application works.

   **Key Features & Their Functions:-**
    
  **Element Panel:**
    * Allows inspection and modification of HTML & CSS in real-time.
    * Helps in testing different styles, layouts, and responsiveness.
    * Useful for debugging issues like incorrect alignment or missing styles.

   **Console Panel:**
    * Displays error messages, warnings, and logs from JavaScript execution.
    * Supports interactive testing of JavaScript commands.
    * Useful for debugging client-side scripts.

   **Network Panel:**
    * Monitors all network requests made by the web page.
    * Shows details like loading time, request headers, and status codes.
    * Helps in optimizing performance by identifying slow-loading resources.

  **Source Panel:**
    * Provides access to website files, including JavaScript, CSS, and source maps.
    * Supports setting breakpoints to pause code execution for debugging.
    * Allows step-by-step execution of JavaScript code.

  **Performance Panel:**
    * Analyzes rendering speed and frame rates.
    * Identifies bottlenecks affecting page performance.
    * Helps optimize animations and transitions.

  **Application Panel:**
    * Displays storage options like Cookies, LocalStorage, and SessionStorage.
    * Shows service workers and cache data for Progressive Web Apps (PWAs).
    * Helps in debugging authentication tokens and user data storage.


#Q-4 What are the advantages of using a Version Control System like Git in a development project?

 * **1. Tracks Changes & History**
    * Git keeps a detailed history of every change made to the project.
    * Developers can view, revert, or restore any previous version of the code.
    * Helps in debugging issues by tracking when and where a bug was introduced.

  **2. Enables Collaboration & Teamwork**
    * Multiple developers can work on the same project simultaneously without conflicts.
    * Distributed version control enables teams across different locations to collaborate efficiently.

  **3. Backup & Recovery**
    * Git serves as a backup system, preventing accidental data loss.
    * If a local machine crashes, the latest code is still available in a remote repository (e.g., GitHub, GitLab, Bitbucket).

   **4. Works Offline (Distributed Version Control)**
    * Unlike centralized VCS (e.g., SVN), Git allows developers to commit and work locally without needing an internet connection.
    * Changes can be pushed to the remote repository later.

   **5. Open-Source & Widely Used**
    * Git is free, open-source, and supported by millions of developers worldwide.
    * Works seamlessly with platforms like GitHub, GitLab, Bitbucket, and Azure DevOps.

#Q-5 Compare and contrast a text editor and an IDE, highlighting their key features and uses?
  
  * **Text Editor:-** A text editor is a lightweight application that allows users to write and edit plain text files, including code. It is typically used for simple coding tasks and offers minimal support for debugging and project management.

   **Key Features of a Text Editor:-**
    * **Lightweight & Fast:-** Uses fewer system resources than an IDE.
    * **Syntax Highlighting:-** Colors different parts of the code for better readability.
    * **Plugins & Extensions:-** Can be extended for extra functionality (e.g., auto-completion, linting).
    * **Multi-language Support:-** Supports various programming languages without requiring heavy configurations.
    * **Basic Code Navigation:-** Provides search and replace funtionality.
    * **Popular Text Editors:-** Notepad++, Sublime Text, Visula studio code.

     
  * **Integrated Development Enviornment(IDE):-** An IDE is a feature-rich software suite designed for full-scale software development. It includes a text editor, compiler, debugger, and other development tools in one package.

   **Key Features of an IDE:-**
    * **Code Compilation & Execution:-** Supports compiling and running code within the environment.
    * **Debugging Tools:-**  Provides breakpoints, watch variables, and stack traces for debugging.
    * **Version Control Integration:-**Works with Git, GitHub, and other version control systems.
    * **Built in Terminal:-** Allows executing commands without leaving the IDE.
    * **Project & Dependency Management:-** Manages large projects efficiently, handling multiple files and dependencies.
    * **Inteligent Code Assistance:-** Auto-completion, refactoring, and syntax error detection.
    * **Popular IDEs:-** Visual Studio (for C#, .NET), JetBrains IntelliJ IDEA (for Java), PyCharm (for Python).
